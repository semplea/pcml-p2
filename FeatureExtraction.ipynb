```python
import numpy as np
import pandas as pd
import pickle

data_folder = 'twitter-datasets/'
```

```python
embeddings = np.load(data_folder + 'embeddings.npy')
embeddings.shape
```

```python
with open(data_folder + 'cooc.pkl', 'rb') as f:
    cooc = pickle.load(f)
cooc_values = np.unique(cooc.row)
len(cooc_values)
cooc_values.shape
```

```python
with open(data_folder + 'vocab.pkl', 'rb') as f:
    vocab = pickle.load(f)
vocab
```

```python
train_set = np.empty([0,embeddings.shape[1]])
label_set = []
label = 1
for fn in ['pos_train.txt', 'neg_train.txt']:
    if fn == 'pos_train.txt':
        label = 1
    else:
        label = -1
    with open(data_folder + fn) as f:
        #iterate over each line of a file
        for line in f:
            score = np.empty([0,embeddings.shape[1]])
            #iterate over each word in the tweet and create a score by taking the embedding of each word
            for token in line.strip().split():
                voc_idx = vocab.get(token, -1)
                if voc_idx >= 0:
                    cooc_idx = np.argwhere(cooc_values == voc_idx).flatten()
                    score = np.append(score, embeddings[cooc_idx], axis=0)
            if len(score) > 0:      
                train_set = np.append(train_set, np.expand_dims(np.mean(score,axis = 0), axis=0), axis = 0)
            else:
                train_set = np.zeros([1,embeddings.shape[1]])
            label_set = np.append(label_set, label)    
np.save(data_folder + 'train_set', train_set)
np.save(data_folder + 'label_set',  label_set)


```
