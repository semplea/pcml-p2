```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_style("whitegrid")
```

```python
glove_dyn = pd.read_csv('graph/run_glove25_dynamic_1482337680-summaries-dev,tag_accuracy_1.csv')
glove_dyn_stat = pd.read_csv('graph/run_glove25_dynamic_static_1482339917-summaries-dev,tag_accuracy_1.csv')
glove_stat = pd.read_csv('graph/run_glove25_static_1482336059-summaries-dev,tag_accuracy_1.csv')
random_dyn = pd.read_csv('graph/run_random_dynamic_1482334802-summaries-dev,tag_accuracy_1.csv')
```

```python
fig, ax = plt.subplots(figsize=(12,9))
glove_dyn.plot(x='Step', y='Value', ax=ax, label='GLoVE dynamic')
glove_dyn_stat.plot(x='Step', y='Value', ax=ax, label='GLoVE dynamic and static')
glove_stat.plot(x='Step', y='Value', ax=ax, label='GLoVE static')
random_dyn.plot(x='Step', y='Value', ax=ax, label='Random dynamic')

ax.legend(
   loc='best',
   fontsize=24)
ax.tick_params(labelsize=18)
plt.xlabel('Step', fontsize=24)
plt.ylabel('Validation accuracy', fontsize=24)
plt.show()
```
